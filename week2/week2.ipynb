Market Share Analysis:

Calculate and compare the market share of each company in different cities or regions.
Analyze market share trends over time.

```python


# Calculate the market share of each company in different cities
market_share_by_city = merged_data.groupby(['City', 'Company'])['Transaction ID'].count().unstack()
total_rides_by_city = market_share_by_city.sum(axis=1)
market_share_by_city = market_share_by_city.div(total_rides_by_city, axis=0) * 100

# Create a bar chart to visualize market share by city
plt.figure(figsize=(12, 6))
sns.barplot(data=market_share_by_city, x=market_share_by_city.index, y='Yellow Cab', color='yellow', label='Yellow Cab')
sns.barplot(data=market_share_by_city, x=market_share_by_city.index, y='Pink Cab', color='pink', label='Pink Cab')
plt.title('Market Share by City')
plt.xlabel('City')
plt.ylabel('Market Share (%)')
plt.xticks(rotation=90)
plt.legend(title='Company')

# Add value labels on the bar chart
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

plt.show()

# Analyze market share trends over time (assuming 'Date of Travel' is the date column)
merged_data['Date of Travel'] = pd.to_datetime(merged_data['Date of Travel'])
merged_data['Year-Month'] = merged_data['Date of Travel'].dt.to_period('M')
market_share_over_time = merged_data.groupby(['Year-Month', 'Company'])['Transaction ID'].count().unstack()
total_rides_over_time = market_share_over_time.sum(axis=1)
market_share_over_time = market_share_over_time.div(total_rides_over_time, axis=0) * 100

## Convert the 'Year-Month' column to a string format (if not already)
market_share_over_time.index = market_share_over_time.index.strftime('%Y-%m')

# Create a line chart to visualize market share trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=market_share_over_time, markers=True)
plt.title('Market Share Trends Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Market Share (%)')
plt.legend(title='Company')
plt.xticks(rotation=45)
plt.show()



```
